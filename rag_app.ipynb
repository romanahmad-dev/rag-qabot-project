```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="google/flan-t5-base")

def qa_bot(pdf_file, query):
    try:
        file_path = pdf_file.name
        loader = PyPDFLoader(file_path)
        documents = loader.load()
        context = " ".join([doc.page_content for doc in documents])

        result = qa_pipeline(question=query, context=context)
        return result['answer']
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
from langchain.document_loaders import TextLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

# Load the document
loader = TextLoader("new-Policies.txt")
documents = loader.load()

# Create embedding model
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create and persist vector database
vectordb = Chroma.from_documents(documents, embedding_model, persist_directory="./chroma_db_retriever")
vectordb.persist()
```

```python
# Use ChromaDB as retriever
retriever = vectordb.as_retriever(search_kwargs={"k": 2})

# Define query
query = "Email policy"

# Perform retrieval
docs = retriever.get_relevant_documents(query)

# Display results
for i, doc in enumerate(docs, 1):
    print(f"Result {i}:\n{doc.page_content}\n{'-'*80}")
```

```python
import gradio as gr
from langchain.document_loaders import PyPDFLoader
from transformers import pipeline

# Load QA Model directly from Hugging Face
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
from transformers import pipeline
summary_pipeline = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_pdf(pdf_file):
    loader = PyPDFLoader(pdf_file.name)
    documents = loader.load()
    text = " ".join([doc.page_content for doc in documents[:3]])
    text = text[:1024]
    summary = summary_pipeline(text)[0]['summary_text']
    return summary
```

```python
# Define QA function
def qa_bot(pdf_file, query):
    try:
        loader = PyPDFLoader(pdf_file.name)
        documents = loader.load()

        # Shorten the context (e.g., just first 3 pages or top 1000 characters)
        context = " ".join([doc.page_content for doc in documents[:3]])
        context = context[:1000]  # limit context length for better model input

        result = qa_pipeline(question=query, context=context)
        return result["answer"]
    except Exception as e:
        return f"Error: {str(e)}"


# Gradio Interface
iface = gr.Interface(
    fn=qa_bot,
    inputs=[
        gr.File(label="Upload PDF"),
        gr.Textbox(lines=2, placeholder="Ask a question...", label="Your Question")
    ],
    outputs="text",
    title="QA Bot - PDF + HuggingFace",
    description="Upload a PDF and ask questions based on its content"
)

iface.launch()
```
